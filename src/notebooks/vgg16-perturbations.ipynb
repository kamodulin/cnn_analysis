```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../data/perturbation-experiments/20210726060901-experiments.csv').applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
node = data[data[' level'] == 'node']
node.pivot(index=" fraction", columns="layer", values=" accuracy").plot(style="-o", figsize=(10, 5), cmap="viridis")
plt.legend(loc='lower left')
plt.xlabel('proportion of nodes knocked out')
plt.ylabel('top-5 cls performance')
plt.title('Node Knockout - VGG16')
```

```python
synapse = data[data[' level'] == 'synapse']
synapse.pivot(index=" fraction", columns="layer", values=" accuracy").plot(style="-o", figsize=(10, 5), cmap="viridis")
plt.legend(loc='lower left')
plt.xlabel('proportion of synapses knocked out')
plt.ylabel('top-5 cls performance')
plt.title('Synapse Knockout - VGG16')
```

```python
run1 = pd.read_csv('../data/perturbation-experiments/20210726060901-experiments.csv').applymap(lambda x: x.strip() if isinstance(x, str) else x)
run2 = pd.read_csv('../data/perturbation-experiments/20210801070321-experiments.csv').applymap(lambda x: x.strip() if isinstance(x, str) else x)
run3 = pd.read_csv('../data/perturbation-experiments/20210801070322-experiments.csv').applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
node = (run1[run1[' level'] == 'node'].pivot(index=" fraction", columns="layer", values=" accuracy") + \
run2[run2[' level'] == 'node'].pivot(index=" fraction", columns="layer", values=" accuracy") + \
run3[run3[' level'] == 'node'].pivot(index=" fraction", columns="layer", values=" accuracy")) / 3
node.plot(style="-o", figsize=(10, 5), cmap="viridis")
plt.legend(loc='lower left')
plt.xlabel('proportion of nodes knocked out')
plt.ylabel('top-5 cls performance')
plt.title('Node Knockout - VGG16')
```

```python
synapse = (run1[run1[' level'] == 'synapse'].pivot(index=" fraction", columns="layer", values=" accuracy") + \
run2[run2[' level'] == 'synapse'].pivot(index=" fraction", columns="layer", values=" accuracy") + \
run3[run3[' level'] == 'synapse'].pivot(index=" fraction", columns="layer", values=" accuracy")) / 3
synapse.plot(style="-o", figsize=(10, 5), cmap="viridis")
plt.legend(loc='lower left')
plt.xlabel('proportion of synapses knocked out')
plt.ylabel('top-5 cls performance')
plt.title('Synapse Knockout - VGG16')
```
